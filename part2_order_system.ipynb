**Task 1 - Explore the Menu**

```python
# Printing the full menu grouped by category

categories = []

for details in menu.values():
    if details["category"] not in categories:
        categories.append(details["category"])

for category in categories:
  print(f"\n{'=' * 5} {category} {'=' * 5}")
  for item_name, details in menu.items():
    if details["category"] == category:
      availability = "[Available]" if details["available"] else "[Unavailable]"
      print(f"{item_name: <18} ₹{details['price']:^6.2f} {availability}")
```

```python
# Using dictionary methods, compute and print:

# Total number of items on the menu

total = len(menu)
print(f"Total number of items on the menu: {total}")

# Total numbers of dishes available in the menu

total_dishes_available = len([dish for dish in menu.values() if dish["available"]])
print(f"\nTotal dishes available: {total_dishes_available}")

# Most expensive item

max_price = max(menu, key=lambda x: menu[x]["price"])
print(f"\nMost expensive item: {max_price} - ₹{menu[max_price]['price']:.2f}")

# Items priced under ₹150

items_under_150 = [item for item, details in menu.items() if details["price"] < 150]
print("\nItems priced under ₹150:")
for item in items_under_150:
  print(f"{item:<18} - ₹{menu[item]['price']:.2f}")
```

**Task 2 — Cart Operations**

```python
cart = []

def display_cart(label=""):
  if label:
    print(f"\n {label}")
  if not cart:
    print("Cart is empty")
    return
  for entry in cart:
    subtotal = entry["quantity"] * entry["price"]
    print(f"{entry['quantity']} x {entry['item']} - ₹{subtotal:.2f}")

# Adding or merging item to cart

def add_to_cart(item, quantity):
  if item in menu and menu[item]["available"]:
    for entry in cart:
      if entry["item"] == item:
        entry["quantity"] += quantity
        print(f"✓ Updated {item} quantity to {entry['quantity']}.")
        return
    cart.append({"item": item, "quantity": quantity, "price": menu[item]["price"]})
    print(f"✓ Added {quantity} x {item} to cart.")

# Show appropriate message for unavailable or non-existing items

  elif item not in menu:
       print(f"✗ {item} is not on the menu.")

  elif not menu[item]["available"]:
       print(f"✗ {item} is currently not available.")

# Removing an item

def remove_from_cart(item, quantity):
    for entry in cart:
      if entry["item"] == item:
        cart.remove(entry)
        print(f"✓ Removed {quantity} x {item} from cart.")
        return
    print(f"✗ {item} is not in the cart.")

# Update quantity of an existing item in the cart

def update_quantity(item, new_quantity):
    for entry in cart:
      if entry["item"] == item:
        entry["quantity"] = new_quantity
        print(f"✓ Updated {item} quantity to {new_quantity}.")
        return
    print(f"✗ {item} is not in the cart.")
```

**Task 3 — Inventory Tracker with Deep Copy**

```python
# Create deep copy

inventory_backup = copy.deepcopy(inventory)

# Manually changing one value in inventory and then print both inventory and inventory_backup

inventory["Paneer Tikka"]["stock"] = 15

print("Inventory after the change:")
print(inventory)
print("\n Inventory_backup remain unchanged")
print(inventory_backup)
```

```python
# Restore original state

inventory["Paneer Tikka"]["stock"] = 10
print("Restored Inventory:")
print(inventory)
```

```python
# Deduct cart quantities from inventory

for entry in cart:
  item = entry["item"]
  quantity_req = entry["quantity"]
  available_stock = inventory[item]["stock"]

  if quantity_req > available_stock:
    print(f"(Insufficient stock for '{item}): "
    f"Required {quantity_req} but only {available_stock} available. Deducting only {available_stock}.")
    inventory[item]["stock"] = 0

  else:
    inventory[item]["stock"] -= quantity_req
    print(f"✓ Deducting {quantity_req} x {item} from inventory. "
    f"Remaining stock: {inventory[item]['stock']}")
```

```python
# Reorder alert

print("\nReorder Alert:")
alerts_found = 0

for item, details in inventory.items():
  if details["stock"] <= details["reorder_level"]:
    print(f"⚠ Reorder Alert: {item} -"
    f"Only {details['stock']} units available "
    f"(reorder level: {details['reorder_level']})")

  if not alerts_found:
    print(f"⚠ Reorder Alert: {item} - "
    f"Only {details['stock']} units available "
    f"(reorder level: {details['reorder_level']})")
```

```python
# Comparing inventory and backup

print("\nInventory Comparison:")
print("\nChanged inventory:")
print(inventory)
print("\nInventory backup (remain unchanged):")
print(inventory_backup)
```

**Task 4 — Daily Sales Log Analysis**

```python
# Daily revenue per day

def revenue_per_day(sales_log):
  return {date: sum(entry["total"] for entry in entries) for date, entries in sales_log.items()}
print(revenue_per_day(sales_log))
```

```python
# Best selling day

def best_selling_day(sales_log):
  return max(sales_log, key=lambda date: sum(entry["total"] for entry in sales_log[date]))

print(f"Best selling day: {best_selling_day(sales_log)}")

```

```python
# Most ordered item

item_counter = {}
for entries in sales_log.values():
  for entry in entries:
    for item in entry["items"]:
      item_counter[item] = item_counter.get(item, 0) + 1

most_ordered_item = max(item_counter, key=item_counter.get)
print(most_ordered_item)
```

```python
# Add new day and reprint the revenue-per-day table and best-selling day to show the stats update

sales_log["2025-01-05"] = [
    {"order_id": 11, "items": ["Butter Chicken", "Gulab Jamun", "Garlic Naan"], "total": 490.0},
    {"order_id": 12, "items": ["Paneer Tikka", "Rasgulla"],                     "total": 260.0},
]

rev_updated = revenue_per_day(sales_log)
print("Revenue Per Day (After adding 2025-01-05):")
print(rev_updated)

best_selling_day_updated = best_selling_day(sales_log)
print("\nBest Selling Day (After adding 2025-01-05)")
print(best_selling_day_updated)
```

```python
# Print all orders (using Enumerate)

all_orders = [
    (date, entry)
    for date in sorted(sales_log.keys())
    for entry in sales_log[date]
]

for num, (date, entry) in enumerate(all_orders, start=1):
  items_str = ", ".join(entry['items'])
  print(f" {num:<4} [{date}] Order #{entry['order_id']:<3}"
  f"- ₹{entry['total']:<8.2f}- Items: {items_str}")
```
